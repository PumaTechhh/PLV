Linear regression by using Deep Neural network: Implement Boston housing price
predictionproblem by Linear regression using Deep Neural network. Use Boston House price predictiondataset.


```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import boston_housing
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 1: Load the dataset
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
# Step 2: Assign column names as in the original dataset
columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"
]

# Step 3: Convert training data to pandas DataFrame
data = pd.DataFrame(x_train, columns=columns)
data['MEDV'] = y_train
```

```python
# Step 4: Display first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Step 5: Check for null values
print("\nChecking for missing values:")
print(data.isnull().sum())
```

```python
# Step 6: Splitting features and target
X_train = x_train
X_test = x_test
y_train = y_train
y_test = y_test
```

```python
# Step 7: Normalize the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 8: Fit a Linear Regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Step 9: Predict on test data
y_pred = model.predict(X_test_scaled)
```

```python
# Step 10: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nMean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# Step 11: Plot Actual vs Predicted
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([0, 50], [0, 50], color='red', linestyle='--')
plt.xlabel("Actual MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Actual vs Predicted MEDV")
plt.grid(True)
plt.show()
```
